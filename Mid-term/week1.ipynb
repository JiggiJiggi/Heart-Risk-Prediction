```python
#new data file
df3 = pd.read_csv('/Users/rathanrajdasari/Downloads/DS&ML/3_transactions.csv') #new data file
df3
```

```python
#merging two dfs
mf = pd.merge(left = df3,right=df2,left_on='customer_id',right_index= True)
```

```python
#max, min, avg amounts spent per hour transaction 
mf.groupby('cust_name').aggregate({'txn_total': ['min','max','mean']})
```

```python
#find the total amnt spent for each banking category
mf.groupby('cust_category').aggregate({'txn_total':'sum'})
```
