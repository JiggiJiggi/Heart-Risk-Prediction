# June 13 Filling null values

```python
#Display the rows that contain at least two missing values
mort[mort.isna().sum(axis=1).eq(2)]
```

```python
#df.dropna() Read the documentation and check out the parameters
.dropna(thresh=2)
#or
mort.dropna(subset=['DeathRate'],inplace=False).head()
```

```python
#mean and median for filling
mort['DeathRate'].describe()
```

```python
#where the death rate is missing: as we can see, the death rate is in the range of 1000, but the median is in the tens, 
#so what we can do is subset the missing values, find the median of it, and fill it
mort[mort['DeathRate'].isna()]
```

```python
#filling null values using median imputation in the DeathRate column
mort['DeathRate'].fillna(value=mort['DeathRate'].median(),inplace=True)
```

```python
mort.isna().sum()
```

```python
#filling missing values in the year column using the forward fill method imputation
mort[mort['Year'].isna()]
```

```python
#using forward fill
mort['Year'].fillna(method = "ffill",limit=2,inplace=True)
mort.isna().sum()
```
