Q3. What is the average spend of customers in Alberta?

```python
df2= pd.read_csv('/Users/rathanrajdasari/Downloads/DS&ML/3_transactions.csv')
mf = pd.merge(left=df2,right=df,left_on='customer_id',right_index=True)
pf = mf.groupby('province')[['txn_total']].mean()

pf.head(1)
```

Q4. How many personal banking customers are there in each province?
